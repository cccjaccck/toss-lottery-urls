```python
import pandas as pd
```

```python
ban_list = [
    'https://toss.im/_m/BUj6u2s?c=573630',
    'https://toss.im/_m/BUj6u2s?c=g1A8Z4'
]
```

```python
def kick_url(array):
    for i in range(len(array) -1, -1, -1):
        if array[i] in ban_list:
            array.pop(i)
```

```python
df = pd.read_csv("./KakaoTalk_Chat_토스 복권 공유방_2024-08-06-14-25-04.csv")
```

```python
filtered_list = []
target_str = "https://toss.im/_m/BUj6u2s?"

def check_word_in(text):
    return target_str in text

for row in df['Message']:
    if check_word_in(row):
        # split sentence
        words = row.split()
        new_words = filter(check_word_in, words)
        filtered_list += new_words

result = list(set(filtered_list))
```

```python
# 공지 손으로 모은 것들..
lines = []
with open("gongji.txt", 'r') as f:
    lines = f.readlines()
#     lines = list(map(lambda x: x[2:-1], lines))
    lines = list(map(lambda x: x.replace('\n',''), lines))
    lines = filter(check_word_in, lines)
result += lines
result = list(set(result))
```

```python
import json
import datetime

kick_url(result)

with open('result.json', 'w') as f:
    now = datetime.datetime.now()
    json.dump({'date': now.strftime('%Y년%m월%d일 %H:%M:%S'), 'result': result}, f)
```
