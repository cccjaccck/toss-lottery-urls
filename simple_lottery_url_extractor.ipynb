```python
import pandas as pd
```

```python
df = pd.read_csv("./KakaoTalk_Chat_toss_lottery_2024-08-06-10-35-03.csv")
```

```python
filtered_list = []
target_str = "https://toss.im/_m/BUj6u2s?"

def check_word_in(text):
    return target_str in text

for row in df['Message']:
    if check_word_in(row):
        # split sentence
        words = row.split()
        new_words = filter(check_word_in, words)
        filtered_list += new_words

result = list(set(filtered_list))
```

```python
len(result), result
```

```python
with open("result.md", 'r') as f:
    lines = f.readlines()
    lines = list(map(lambda x: x[2:-1], lines))
result += lines
result = list(set(result))
```

```python
import json
import datetime

with open('result.json', 'w') as f:
    now = datetime.datetime.now()
    json.dump({'date': now.strftime('%Y년%m월%d일 %H:%M:%S'), 'result': result}, f)
```

```python
with open("result.md", 'w') as f:
    for url in result:
        f.write(f'- {url}\n')
```
